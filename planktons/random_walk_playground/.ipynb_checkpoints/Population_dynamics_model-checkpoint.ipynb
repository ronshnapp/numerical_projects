# population distribution as a competition between turbulence and encounters


__A general question:__ How can a population of plankton keep beeing close together in a vast and turbulent ocean?


__A toy model:__ plankton are sweeming randomly in space; we assume they can detect their neighbours if they are found within some perception sphere. The plankton are assumed to all swim with the same speed $V$ but at random directions. We also assume that the plankton have a social memory of duration $M$ time units, so they will only want to encounter a new plankter if they hadn't met another plankter for that duration.

```python
class plankter(object):
    '''
    A class that holds the position and velocity
    of each plankter in the simulation
    
    p0 - initial velocity
    V - swimming speed
    dt - time step
    encounters - a list of values, =0 if there was no encounter in this time step and =1 if there was.
    '''
    
    def __init__(self, p0, V, dt):
        
        self.dt = dt
        self.V = V
        self.p = [p0]
        self.v = []
        self.encounters = [0]
        
        self.generate_new_velocity()
        
        
    def generate_new_velocity(self):
        
        r = np.random.uniform(-1,1,2)
        self.v.append( r / np.linalg.norm(r) * self.V )
    
    
    def step(self):
        self.p.append( self.p[-1] + self.v[-1] * self.dt )
        self.generate_new_velocity()
        self.encounters.append(0)
```

```python
class ocean(object):
    '''
    a class that holds the entire simulation
    planktor - a list of plankter objects
    R - the perception radius
    '''
    
    def __init__(self, plankton_lst, R, M):
        self.plankton = plankton_lst
        self.R = R
        self.M = M
        self.dt = self.plankton[0].dt
        
    
    def step(self):
        
        # first, advance all plankters one time step:
        for p in self.plankton: p.step()
           
        # get indexes of pairs that make encounters:
        pairs_indexes = self.find_close_plankters()
        N_pairs = len(pairs_indexes[0])
        
        # make the pair meet by changing their velocity, and mark the encounter:
        for k in range(N_pairs):
            i,j = pairs_indexes[0][k], pairs_indexes[1][k]
            pi, pj = self.plankton[i], self.plankton[j]
            meeting_point = (pi.p[-1] + pj.p[-1])/2
            vi = (meeting_point - pi.p[-1])/self.dt
            vj = (meeting_point - pj.p[-1])/self.dt
            pi.v[-1] = vi
            pj.v[-1] = vj
            pi.encounters[-1] = 1
            pj.encounters[-1] = 1
    
    
    def find_close_plankters(self):
        '''
        returns indexes of pair pf plankton that should bepaired in the next step
        '''
        N = len(self.plankton)
        dM = int(self.M / self.dt)
        encounter_matrix = np.ones((N, N)) * -1
        for i in range(N):
            for j in range(i+1, N):
                d = self.distance(self.plankton[i], self.plankton[j])
                social_index_i = 1 in self.plankton[i].encounters[-dM:] 
                social_index_j = 1 in self.plankton[j].encounters[-dM:]
                
                # both plankters are not ready for an encounter:
                if social_index_i or social_index_j: 
                    encounter_matrix[i,j] = -1
                    continue
                  
                # if plankters are too far away:
                elif d>self.R: 
                    encounter_matrix[i,j] = -1
                    continue
                
                else:
                    encounter_matrix[i,j] = d
                    
        # in each row and column keep only the pair minimum distance plankton:
        for i in range(N):
            w = np.where(encounter_matrix[i,:]>0)[0]
            if len(w)>1:
                mn = np.amin(encounter_matrix[i,:][w])
                encounter_matrix[i,:][encounter_matrix[i,:]>mn] = -1
                        
        for j in range(N):
            w = np.where(encounter_matrix[:,j]>0)[0]
            if len(w)>1:
                mn = np.amin(encounter_matrix[:,j][w])
                encounter_matrix[:,j][encounter_matrix[:,j]>mn] = -1
        
        return np.where(encounter_matrix > 0)
    
    
    def distance(self, plankter_i, plankter_j):
        '''returns the distance between a pair of plankters'''
        return np.linalg.norm(plankter_i.p[-1] - plankter_j.p[-1])
    
    
    def plot_at_step(self, step_i, fig, ax):
        '''will plot as a scatter plot the position of plankters at time step_i'''
        dM = int(self.M/self.dt)
        for i in range(len(self.plankton)):
            p = self.plankton[i]
            if 1 in p.encounters[step_i - dM:step_i]: c = 'y'
            else: c = 'r'
            ax.plot(p.p[step_i][0], p.p[step_i][1], 'o', ms=3, color=c)
            
    
    def calc_bbox(self):
        '''determines the boundaries of the plankton trajectories'''
        x_lst = []
        y_lst = []
        for p in self.plankton:
            x_lst += list(np.array(p.p)[:,0])
            y_lst += list(np.array(p.p)[:,1])
            
        self.bbox = (np.amin(x_lst), np.amax(x_lst), np.amin(y_lst), np.amax(y_lst))
        
    
    def get_distance_from_center_at_step_i(self, step_i):
        '''will return a list of the distances of all
        plankters from their center of the distribution at
        time step_i'''
        
        x_lst = []
        y_lst = []
        
        for p in self.plankton:
            x_lst.append(p.p[step_i][0])
            y_lst.append(p.p[step_i][1])
        
        cx = np.mean(x_lst)
        cy = np.mean(y_lst)
        
        return ((np.array(x_lst) - cx)**2 + (np.array(y_lst) - cy)**2)**0.5
        
        
    def get_distance_between_neighbours_at_step_i(self, step_i):
        '''will return a list of the distances between plankton at
        time step_i'''
        d_lst = []
        N = len(self.plankton)
        for i in range(N):
            for j in range(i+1, N):
                r = self.plankton[i].p[step_i] - self.plankton[j].p[step_i]
                d_lst.append(np.linalg.norm(r))
        return d_lst
        
```

# Some theoretical considerations:

#### 1. It is just a random walk model for _non-interacting_ planktons:


$x_{i+1} = x_i + V\, dt\, \cos(\theta_i)$


$y_{i+1} = y_i + V\, dt\, \sin(\theta_i)$


$ \langle r_i^2 \rangle = i \, V^2 \, dt^2$



#### 2. The parameters of the model:

1. perception radius $R$
2. memory time $M$
3. velocity $V$ and timestep $dt$ 

Then, the length of steps in each step is $S = V dt$. Also, the squared average distance of travel within the period of the memory duration is equal to: $S_M^2 = \frac{M}{dt}V^2\,dt^2 = M \, V^2 \, dt$


Finally, the parametrs can be non-dimentionalized as follows: 

1. We normalize the perception raduis using the step size: $$\rho \equiv \frac{R}{S} = \frac{R}{V\, dt}$$ 
2. A dimensionless memory kernel: $$\mu \equiv \frac{S_M}{R} = \sqrt{\frac{M \, dt \,V^2}{R^2}}$$


Note that $\mu$ is a very important parameter. If $\mu \ll 1$ ($R\gg S_M$), then a pair that first interacted can hardly ever escape. If $\mu \gg 1$ ($R \ll S_M$) then pair will almost always escape each other after interacting. Therefore, $\mu$ resembels an inverse of an interaction strength parameter.

## Experiments in a loop:

```python
import time

Np = 100      # <--- number of plankton
V = 1.0       # <--- jump velocity
dt = 1.0      # <--- jump waiting time
R = 10.0       # <--- encounter raduis
M = dt * 5    # <--- encounter memory
R_init = 0.25

N_steps = 200
N_repets = 10 # number of experiment repetitions



# run experiment in a loop:
for j in range(N_repets):
    t0=time.time()
    
    # initiate and run:
    p_lst = [plankter(np.random.normal(0,R_init,2), V, dt) for i in range(Np)]
    O = ocean(p_lst, R, M)
    for i in range(N_steps):
        O.step()
    
    # initiate and run:
    positions = []
    for pln in O.plankton:
        positions.append( np.array(pln.p) )

    positions = np.array(positions)
    np.savez('run10_%d.npz'%(j+10), positions  = positions)

    print(j, '%.2f'%(time.time() - t0))
```

# A periodic ocean model:

same as the Ocean above, only here, the box is periodic, i.e. a torus.



Recall the dimensionless parameters:

1. $$\rho \equiv \frac{R}{S} = \frac{R}{V\, dt}$$  
2. $$\mu \equiv \frac{S_M}{R} = \sqrt{\frac{M \, dt \,V^2}{R^2}}$$

Adding a bounded domain size adds new variables to consider: the domain size $L$ and the number of cells, $N$. We can normalize $L$ in several different ways:

1. using the step size: $$\lambda_S = \frac{S}{L} = \frac{V \, dt}{L}$$
2. using the perception radius: $$ \lambda_R = \frac{R}{L} $$

For the simulation to be interesting, both $\lambda_S$ and $\lambda_R$ should be rather small. Another interesting parameter to consider is the normalized density of the planktons in the domain: $$ c = N \frac{\pi R^2}{L^2}$$





```python
class ocean(object):
    '''
    a class that holds the entire simulation
    
    plankton_lst - a list of plankter objects
    R - the perception radius
    M - the memory length of previous encounters
    L - Domain box size. If L is a number the the domain is between
        the boundaries (0,L) in each axis and they are periodic
        boundaries. If L is None (default) then the domain is infinite.
    '''
    
    def __init__(self, plankton_lst, R, M, L=None):
        self.plankton = plankton_lst
        self.R = R
        self.M = M
        self.dt = self.plankton[0].dt
        
        if L is None:
            self.boundedDomain = False
        
        else:
            self.boundedDomain = True
            self.L = L
        
    
    def step(self):
        
        # first, advance all plankters one time step:
        for p in self.plankton: p.step()
           
        # get indexes of pairs that make encounters:
        pairs_indexes = self.find_close_plankters()
        N_pairs = len(pairs_indexes[0])
        
        # make the pair meet by changing their velocity, and mark the encounter:
        for k in range(N_pairs):
            i,j = pairs_indexes[0][k], pairs_indexes[1][k]
            pi, pj = self.plankton[i], self.plankton[j]
            meeting_point = (pi.p[-1] + pj.p[-1])/2
            vi = (meeting_point - pi.p[-1])/self.dt
            vj = (meeting_point - pj.p[-1])/self.dt
            pi.v[-1] = vi
            pj.v[-1] = vj
            pi.encounters[-1] = 1
            pj.encounters[-1] = 1
    
        # if bounded domain reflect all the plankers that
        # are out of (0,L):
        if self.boundedDomain:
            for plankter in self.plankton:
                plankter.p[-1] = plankter.p[-1]%L
            
    
    def find_close_plankters(self):
        '''
        returns indexes of plankton pairs that should bepaired in the next step
        '''
        N = len(self.plankton)
        dM = int(self.M / self.dt)
        encounter_matrix = np.ones((N, N)) * -1
        for i in range(N):
            for j in range(i+1, N):
                d = self.distance(self.plankton[i], self.plankton[j])
                social_index_i = 1 in self.plankton[i].encounters[-dM:] 
                social_index_j = 1 in self.plankton[j].encounters[-dM:]
                
                # both plankters are not ready for an encounter:
                if social_index_i or social_index_j: 
                    encounter_matrix[i,j] = -1
                    continue
                  
                # if plankters are too far away:
                elif d>self.R: 
                    encounter_matrix[i,j] = -1
                    continue
                
                else:
                    encounter_matrix[i,j] = d
                    
        # in each row and column keep only the pair minimum distance plankton:
        for i in range(N):
            w = np.where(encounter_matrix[i,:]>0)[0]
            if len(w)>1:
                mn = np.amin(encounter_matrix[i,:][w])
                encounter_matrix[i,:][encounter_matrix[i,:]>mn] = -1
                        
        for j in range(N):
            w = np.where(encounter_matrix[:,j]>0)[0]
            if len(w)>1:
                mn = np.amin(encounter_matrix[:,j][w])
                encounter_matrix[:,j][encounter_matrix[:,j]>mn] = -1
        
        return np.where(encounter_matrix > 0)
    
    
    def distance(self, plankter_i, plankter_j):
        '''returns the distance between a pair of plankters'''
        return np.linalg.norm(plankter_i.p[-1] - plankter_j.p[-1])
    
    
    def plot_at_step(self, step_i, fig, ax):
        '''will plot as a scatter plot the position of plankters at time step_i'''
        dM = int(self.M/self.dt)
        for i in range(len(self.plankton)):
            p = self.plankton[i]
            if 1 in p.encounters[step_i - dM:step_i]: c = 'y'
            else: c = 'r'
            ax.plot(p.p[step_i][0], p.p[step_i][1], 'o', ms=3, color=c)
            
    
    def calc_bbox(self):
        '''determines the boundaries of the plankton trajectories'''
        x_lst = []
        y_lst = []
        for p in self.plankton:
            x_lst += list(np.array(p.p)[:,0])
            y_lst += list(np.array(p.p)[:,1])
            
        self.bbox = (np.amin(x_lst), np.amax(x_lst), np.amin(y_lst), np.amax(y_lst))
        
    
    def get_distance_from_center_at_step_i(self, step_i):
        '''will return a list of the distances of all
        plankters from their center of the distribution at
        time step_i'''
        
        x_lst = []
        y_lst = []
        
        for p in self.plankton:
            x_lst.append(p.p[step_i][0])
            y_lst.append(p.p[step_i][1])
        
        cx = np.mean(x_lst)
        cy = np.mean(y_lst)
        
        return ((np.array(x_lst) - cx)**2 + (np.array(y_lst) - cy)**2)**0.5
        
        
    def get_distance_between_neighbours_at_step_i(self, step_i):
        '''will return a list of the distances between plankton at
        time step_i'''
        d_lst = []
        N = len(self.plankton)
        for i in range(N):
            for j in range(i+1, N):
                r = self.plankton[i].p[step_i] - self.plankton[j].p[step_i]
                d_lst.append(np.linalg.norm(r))
        return d_lst
        
```

##### nearest neighbour distances: 

```python

def get_nearest_neighbours_list(positions):
    
    N = len(positions)
    
    # make a distances matrix:
    distanceMatrix = np.ones((N,N))*-1

    for j in range(N):
        for jj in range(N):
            if j == jj: continue
            distanceMatrix[j,jj] = np.linalg.norm(positions[j] - positions[jj])
        
    # find minimum distances across each row:
    nearesDistances_lst = []

    for j in range(N):
        lst = distanceMatrix[j,:]
        nearesDistances_lst.append( np.amin(lst[lst>0]) )
        
    return nearesDistances_lst
    
```

##### Radial distribution function:

```python
def get_radial_distribution_function(positions, bins=10):
    
    N = len(positions)
    
    # make a distances matrix:
    distances = []

    for j in range(N):
        for jj in range(j+1, N):
            distances.append(np.linalg.norm(positions[j] - positions[jj]))
    
    h = np.histogram(distances, density=True, bins=bins)
    x_, y_ = 0.5*(h[1][:-1]+h[1][1:]), h[0]
    return x_, y_




fig, ax = plt.subplots()

x, y = get_radial_distribution_function(positions[0], bins=np.linspace(0,1.4,num=150))
ax.plot(x, y)


x, y = get_radial_distribution_function(positions[3000], bins=np.linspace(0,1.4,num=150))
ax.semilogx(x, y, 'o-', ms=2)
```
